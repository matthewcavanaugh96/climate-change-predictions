# Surface Temperature Change

```python
tempchange = pd.read_csv("/Users/matthewcavanaugh/Desktop/Data analysis work/Data projects/Using machine learning to predict climate change/Sources/Original sources/Surface temperature change.csv")
tempchange
```

We can drop most of these columns, as many of them are merely describing the same information, but we do need to keep 'ISO3' as these standardized codes will be important in helping us match country names across all datasets. It will be our stand-in for country names for the time being.

```python
tempchange.drop('ObjectId', axis=1, inplace=True)
tempchange.drop('Country', axis=1, inplace=True)
tempchange.drop('ISO2', axis=1, inplace=True)
tempchange.drop('Unit', axis=1, inplace=True)
tempchange.drop('Source', axis=1, inplace=True)
tempchange.drop('Indicator', axis=1, inplace=True)
tempchange.drop('CTS_Code', axis=1, inplace=True)
tempchange.drop('CTS_Name', axis=1, inplace=True)
tempchange.drop('CTS_Full_Descriptor', axis=1, inplace=True)
tempchange
```

Our year names each have an "F" in front. Let's get rid of that.

```python
tempchange.columns = tempchange.columns.str.replace("^F", "", regex=True)
tempchange.head()
```

Next we will have to remove all columns before 1992 and after 2019, as that is the intersection of years among all our sources.

```python
tempchange = tempchange[['ISO3'] + [str(year) for year in range(1992, 2020)]]
tempchange.head()
```

This dataset has a substantial number of nulls, as reporting is inconsistent by nation. To handle this, we will drop any country('ISO3') with more than five null years. Then, any country with five or fewer nulls will have those nulls filled with the mean. This is intended to balance accuracy while not dropping too many countries.

```python
# Remove countries with more than five nulls
tempchange = tempchange[~tempchange["ISO3"].isin(tempchange.loc[(tempchange.isnull().sum(axis=1)) > 5, "ISO3"])]

# For the remaining countries, fill nulls with the country's mean
tempchange.iloc[:, 1:] = tempchange.iloc[:, 1:].apply(lambda row: row.fillna(row.median()), axis=1)
```

```python
tempchange
```

Here, we will use the package country_converter to convert the ISO3 codes back into standard short country names.

```python
import country_converter as cc

# Convert the ISO3 codes to standard short country names (e.g. USA = United States)
tempchange['ISO3'] = tempchange['ISO3'].apply(lambda x: cc.convert(names=x, to='name_short'))

# Any unmatched ISO3 codes will be rendered as 'not found'
# Warning messages will be displayed below
# Remove these values from the dataframe
tempchange = tempchange[tempchange['ISO3'] != 'not found']

# Change the name of the column to 'Country']
tempchange.rename(columns={'ISO3': 'Country'}, inplace=True)
```

```python
tempchange
```

Our final step is to melt our table to short format, so that 'Year' is a single column, and the temperature information is another.

```python
# Melt the DataFrame
tempchange = tempchange.melt(id_vars=["Country"], 
                                  var_name="Year", 
                                  value_name="Temperature deviation from 1951-80 norms (°C)")

# Sort by Country, then by year
tempchange = tempchange.sort_values(by=["Country", "Year"])

# Display the DataFrame to ensure we've successfully applied our changes
tempchange
```

With that, our surface temperature data is ready! One down, four to go. Let's save it as a CSV so we can load it later.

```python
# First, find the common countries across all five DataFrames
common_countries = set(tempchange['Country']) & set(forestandcarbon['Country']) & set(landcover['Country']) & set(population['Country']) & set(gdp['Country']) & set (emissions['Country'])

# Filter each DataFrames to keep only the common countries
tempchange_filtered = tempchange[tempchange['Country'].isin(common_countries)]
forestandcarbon_filtered = forestandcarbon[forestandcarbon['Country'].isin(common_countries)]
landcover_filtered = landcover[landcover['Country'].isin(common_countries)]
population_filtered = population[population['Country'].isin(common_countries)]
gdp_filtered = gdp[gdp['Country'].isin(common_countries)]
emissions_filtered = emissions[emissions['Country'].isin(common_countries)]

# Merge all DataFrames on 'Country' and 'Year'
climate = tempchange_filtered.merge(forestandcarbon_filtered, on=['Country', 'Year'], how='inner')\
                                 .merge(landcover_filtered, on=['Country', 'Year'], how='inner')\
                                 .merge(population_filtered, on=['Country', 'Year'], how='inner')\
                                 .merge(gdp_filtered, on=['Country', 'Year'], how='inner')\
                                 .merge(emissions_filtered, on=['Country', 'Year'], how='inner')

# Display our combined DataFrame to ensure these changes were properly applied
climate
```

# Recalculate some columns, create new ones, and reorder the columns in a more pleasing way

All columns except "Country" should be in numeric format.

```python
# Convert all columns except 'Country' to numeric, coercing errors to NaN
climate.loc[:, climate.columns != 'Country'] = climate.loc[:, climate.columns != 'Country'].apply(pd.to_numeric, errors='coerce')
```

A few columns are measured by raw land area, which isn't a fair comparison as countries are differently sized. We will recalculate them as percentages. Furthermore, population columns are rendered in thousands, so we will change them to absolute numbers. Finally, we will drop some columns that may be superfluous or redundant.

```python
# Create a new column calculating inland water bodies as a percentage
climate['Inland water bodies (% of total land area)'] = (
    (climate['Inland water bodies (1000 HA)'] / climate['Land area (1000 HA)']) * 100
)
# Drop the original column
climate.drop('Inland water bodies (1000 HA)', axis=1, inplace=True)


# Create a new column calculating permanent snow and glaciers as a percentage
climate['Permanent snow and glaciers (% of total land area)'] = (
    (climate['Permanent snow and glaciers (1000 HA)'] / climate['Land area (1000 HA)']) * 100
)
# Drop the original column
climate.drop('Permanent snow and glaciers (1000 HA)', axis=1, inplace=True)


# Create a new column calculating artificial surfaces as a percentage
climate['Artificial surfaces (% of total land area)'] = (
    (climate['Artificial surfaces (1000 HA)'] / climate['Land area (1000 HA)']) * 100
)
# Drop the original column
climate.drop('Artificial surfaces (1000 HA)', axis=1, inplace=True)


# Create a new column calculating barren land as a percentage
climate['Barren land (% of total land area)'] = (
    (climate['Terrestrial barren land (1000 HA)'] / climate['Land area (1000 HA)']) * 100
)
# Drop the original column
climate.drop('Terrestrial barren land (1000 HA)', axis=1, inplace=True)


# Create a new column calculating births by women 15 to 19 as a percentage
climate['Percent of births by women 15 to 19'] = (
    (climate['Births by women aged 15 to 19 (thousands)'] / climate['Births (thousands)']) * 100
)
# Drop the original column
climate.drop('Births by women aged 15 to 19 (thousands)', axis=1, inplace=True)

# Recalculate population (thousands) to reflect absolute numbers and change the name
climate['Population (thousands)'] = climate['Population (thousands)'] * 1000
climate.rename(columns={'Population (thousands)': 'Population'}, inplace=True)

# Recalculate population density to reflect absolute numbers (name already correct) 
climate['Population density (sq km)'] = climate['Population density (sq km)'] * 1000

# Drop excess columns
climate.drop('Population Change (thousands)', axis=1, inplace=True)
climate.drop('Male population (thousands)', axis=1, inplace=True)
climate.drop('Female population (thousands)', axis=1, inplace=True)
climate.drop('Births (thousands)', axis=1, inplace=True)
```

```python
climate
```

Let's calculate a few new columns. First we will use population and GDP to determine GDP per capita by country for each year. Then we will measure the same relative to 1992. The latter will allow us to see which countries have gained or declined in GDP per capita.

```python
# Calculate GDP per capita
climate['GDP per capita'] = climate['Adjusted GDP in USD'] / climate['Population']

# Measure each country's per capita GDP by year as a percentage of its 1992 value
climate['GDP per capita (% of 1992)'] = climate.groupby('Country')['GDP per capita'].transform(lambda x: (x / x.iloc[0]) * 100)
```

Below is the most important new metric we will be calculating, called ["Warming trend since 1992 (°C)"], as it will be our target in the project notebook. It is calculated using an expanding mean of "Temperature deviation from 1951-80 norms (°C)". Every year's warming trend is the average of yearly temperature variations up to and including that year, starting from 1992. For instance, 2002's warming trend is the average of all temperature variations from 1992 to 2002. This metric attempts to capture a long-term trend in temperature changes and flatten outliers.

We are using this metric as it would be unfair and misleading to simply predict the temperature in a given year, as a country may have had unusually cold temperatures that year despite a warming trend overall. When making future predictions, this will show us how much warmer a future year is expected to be *on average* compared to the baseline.

```python
climate["Warming trend since 1992 (°C)"] = (
    climate.groupby("Country")["Temperature deviation from 1951-80 norms (°C)"]
    .expanding()
    .mean()
    .reset_index(level=0, drop=True)  # Reset index after expanding
)

climate
```
